### GSM8K

```python
from datasets import load_dataset
# load gsm8k dataset
gsm8k_dataset = load_dataset("gsm8k", "main")
gsm8k_dataset
```

```python
lst = []
for i in range(250):
    data_pair = {
        'question': gsm8k_dataset['train'][i]['question'],
        'answer': gsm8k_dataset['train'][i]["answer"].split("####")[-1].strip().lower()
    }
    lst.append(data_pair)
lst
```

```python
import json

with open('dataset.json', 'w') as f:
    json.dump(lst, f)
    
with open('dataset.json', 'r') as f:
    data = json.load(f)
    
print(data)
```

### PubMedQA

```python
from datasets import load_dataset

ds = load_dataset("qiaojin/PubMedQA", "pqa_labeled")
```

```python
ds["train"][0]
```

```python
lst = []

for i in range(250):
    abstract = ds["train"][i]["context"]["contexts"]
    abstract = "\n".join(abstract)
    data_pair = {
        "question": f"""Question: {ds["train"][i]["question"]}\nAbstract: {abstract}""",
        "answer": ds["train"][i]["final_decision"]
    }
    lst.append(data_pair)
    
lst
```

```python
import json

with open('dataset.json', 'w') as f:
    json.dump(lst, f)
    
with open('dataset.json', 'r') as f:
    data = json.load(f)
    
print(data)
```
