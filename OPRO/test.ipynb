```python
modelfile='''
FROM llama3
PARAMETER temperature 1.0
'''
ollama.create(model='llama3_TEMP1', modelfile=modelfile)
```

```python
from token_count import TokenCount
tc = TokenCount(model_name="gpt-4-turbo")
text = "Your text here"
tokens = tc.num_tokens_from_string(text)
print(f"Tokens in the string: {tokens}")
```

```python
tokens = {"input": 0, "output": 0}

def num_tokens_from_string(text):
    tokens["input"] += tc.num_tokens_from_string(text)

num_tokens_from_string("Your text here")
print(f"Tokens in the string: {tokens}")
```
